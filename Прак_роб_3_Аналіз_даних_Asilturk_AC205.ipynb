```python
import numpy as np
#--------------1------------
odessa_months = 10
odessa_cities = 11

kherson_cities = 11*10
kherson_months = 12

odessa_array = np.random.randint(100, 250 , size=[odessa_months, odessa_cities])
kherson_array = np.random.randint(90, 240 , size=[kherson_cities, kherson_months])
print("Odessa:\n", odessa_array, "\n\nKherson:\n", kherson_array)

#--------------2------------
print("\n\nTask 2:")
odessa_kherson_array = np.concatenate((np.transpose(odessa_array[6:8, :]), (kherson_array[:, 6:8])))
print("\n\nConcatenated array:\n",odessa_kherson_array)

#--------------3------------
print("\n\nTask 3:")
min_roaming_duration = np.min(odessa_kherson_array ,axis=0)
max_roaming_duration = np.max(odessa_kherson_array, axis=0)
total_roaming_duration  = np.sum(odessa_kherson_array, axis=0)
average_roaming_duration = np.mean(odessa_kherson_array, axis=1)
print("\nMinimum roaming duration:", min_roaming_duration ,"\nMaximum roaming duration:",max_roaming_duration,"\n\nAverage roaming duration:\n",average_roaming_duration,"\n\nTotal roaming_duration:",total_roaming_duration)

#--------------4------------
print("\nTask 4:")
total_last_3months = np.sum(odessa_kherson_array[6:8, -3:])
print("\nTotal roaming duration for last 3 months: ", total_last_3months)

#--------------5------------
min_value = odessa_kherson_array.min()
max_value = odessa_kherson_array.max()
new_array = np.random.randint(min_value, max_value , size=[10,22])
print("\n\nNew array for task 5:\n",new_array)

#--------------6------------
new_array2 = np.concatenate((odessa_kherson_array, new_array[::,6:8]))
print("\n\nNew array for task 6:\n",new_array2)

#--------------7------------
min_newArray2 = np.min(new_array2, axis=0)
max_newArray2 = np.max(new_array2, axis=0)
print("\n\nMax-Min arrays for task 7:\n\n", "min:\n",np.min(new_array2, axis=0), "\nmax:\n",np.max(new_array2, axis=0), "\nmin:\n",np.min(new_array2, axis=1), "\nmax:\n",np.max(new_array2, axis=1))

months = ['July', 'August']

i=0
for roaming_durations in new_array2:
  min_month_index = np.where(roaming_durations == np.min(roaming_durations,axis=0))
  max_month_index = np.where(roaming_durations == np.max(roaming_durations,axis=0))

  min_roaming_duration = np.min(roaming_durations,axis=0)
  max_roaming_duration = np.max(roaming_durations,axis=0)
  print("In city #",(i+1)," the lowest roaming duration is",min_roaming_duration,"minutes (",str(months[int(min_month_index[0])]),") and the highest roaming duration is",max_roaming_duration,"minutes (",str(months[int(max_month_index[0])]),").")
  i=i+1

#--------------8------------
print("\nTask 8:")
for x in range(new_array2.shape[1]):
  print("\naverage duration of roaming calls of %d months: %d +\- %d" %(x+1, new_array2.mean(axis=0)[x], new_array2.std(axis=0)[x]))

#--------------9------------
print("\n\nResults of task 9:\n")

arr_L = np.percentile(new_array2,60)
print(arr_L,"\n")
arr_L = np.percentile(new_array2, 60, axis=0)
print(arr_L,"\n")
arr_L = np.percentile(new_array2, 60, axis=1)
arr_L,"\n"
```
