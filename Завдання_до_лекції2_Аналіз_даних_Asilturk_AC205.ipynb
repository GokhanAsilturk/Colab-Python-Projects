## Вправа 4

Створіть основний словник із іменами продавців та моделями iPhone. Створіть додатковий словник що містить назву моделі та ціну. Вивести імена тих продавців, хто    продав моделі старше 6 з ціною, що не перевищує $500.

```python
sellers_models = {
    "Apple Offical": ['7','8','X'],
    "rozetka.com" : ['6','7','7 Plus','8','8 Plus','X','11'],
    "citrus.com" : ['6s','11','12'],
    "amazon.com" : ['5','8','13 Pro Max']
}


models_prices = {
    '5' : 200,
    '6' : 300,
    '6s' : 350,
    '7' : 400,
    '7 Plus' : 480,
    '8' : 550,
    '8 Plus' : 620,
    'X' : 700,
    '11' : 800,
    '12' : 930,
    '13 Pro Max' : 1100,
    
}



def show_phones(out_of_model,target_price):
  print("\tResults of Search\n\n")
  for seller, models in sellers_models.items(): 
    stock_info = ""
    
    
    for model in models:
      
      
      for model_type, price in models_prices.items():
        if model_type==model and check_price(price,target_price) and check_model(model,out_of_model):
          stock_info = stock_info + "\n" + ('Iphone %s : %d$' % (model,price)) #burada hata var!

    if stock_info != "":
      print("Seller: %s" %seller)
      print(stock_info + "\n\n")
          
            

    print('\n')
  print("filtered -> lowest model: %s+ , max price: %d$" % (out_of_model,target_price))
      

def check_price(price,target_price):

  if price <= target_price:
    return True
  else:
    return False


def check_model(model,out_of_model):

  if model != out_of_model:
    return True
  else:
    return False



show_phones('5',500)

  


```
