```python
import numpy as np
import csv
import pandas as pd
import io


url = 'https://raw.githubusercontent.com/GokhanAsilturk/game_sales/main/game_sales.csv%20-%20game_sales.csv'

original_dtype = np.dtype([('name', 'S50'), ('platform', 'S10'), ('na_Sales', np.float64), ('eu_Sales', np.float64), ('jp_Sales', np.float64), ('other_Sales', np.float64), ('critic_score', np.int32)])
dtype1 = np.dtype([('name', 'S50'), ('na_Sales', np.float64), ('eu_Sales', np.float64), ('jp_Sales', np.float64), ('other_Sales', np.float64), ('critic_score', np.int32)] )
dtype_platforms = np.dtype([('name', 'S50'), ('platform', 'S10')])

game_sales = np.genfromtxt(url, dtype=original_dtype, usecols=(0,1,5,6,7,8,10), delimiter=',', skip_header=1,filling_values=0)
#print(game_sales)

games1 = np.genfromtxt(url, dtype=dtype1, usecols=(0,5,6,7,8,10), delimiter=',', skip_header=1, max_rows=1001, filling_values=0)
games2 = np.genfromtxt(url, dtype=dtype1, usecols=(0,5,6,7,8,10), delimiter=',', skip_header=1001, filling_values=0)
#print(games1)
#print(games2)

game_names1 = np.loadtxt(url, dtype=dtype_platforms, usecols=(0,1), delimiter=',', skiprows=1, max_rows=1000)
game_names2 = np.loadtxt(url, dtype=dtype_platforms, usecols=(0,1), delimiter=',', skiprows=1000)
#print(game_names1)
#print(game_names2)


#____________________________Task1______________________________________________

game_sales = game_sales[:len(games1)-1]
print("\n\n-----------------Task1------------------------\n\n",len(game_sales))




#____________________________Task2______________________________________________
#Я вже зробив про оцінку критиків "critic_score" ціле число.

#print(dtype1['critic_score'])
games1['critic_score'].astype('int32')
games2['critic_score'].astype('int32')

print("\n\n-----------------Task2------------------------\n\n",games2['critic_score'].dtype)




#____________________________Task3______________________________________________

def countDistinct(arr, n):
    res = 1
    for i in range(1, n):
        j = 0
        for j in range(i):
            if (arr[i] == arr[j]):
                break
        if (i == j + 1):
            res += 1
    return res

print("\n\n-----------------Task3------------------------\n\nUnique names of the Game Platform 1: ",countDistinct(game_sales['platform'],len(game_names1['platform'])))
print("Unique names of the Game Platform 2: ",countDistinct(game_sales['platform'],len(game_names2['platform'])))




#____________________________Task4______________________________________________

def findSamePlatforms(arr1,arr2, n1): 
    platforms = []
    for i in range(n1):
        
        if (arr1[i] in arr2 and arr1[i] not in platforms):
            platforms.append(arr1[i]) 
    return platforms

print("\n\n-----------------Task4------------------------\n\nSame platforms: ",findSamePlatforms(game_names1['platform'],game_names2['platform'],len(game_names1['platform'])))






#____________________________Task5______________________________________________

def findDifferentPlatforms(arr1,arr2, n1, n2): 
    platforms = []
    for i in range(n1):
        if (arr1[i] not in arr2 and arr1[i] not in platforms):
            platforms.append(arr1[i]) 

    for i in range(n2):
        if (arr2[i] not in arr1 and arr2[i] not in platforms):
            platforms.append(arr2[i]) 

    return platforms

print("\n\n-----------------Task5------------------------\n\nDifferent platforms: ",findDifferentPlatforms(game_names1['platform'], game_names2['platform'], len(game_names1['platform']), len(game_names2['platform'])))





#____________________________Task 6 & 7_________________________________________


g_price1 = np.array([100.5, 98.3, 124.4])
g_price2 = np.array([110.7, 99.6, 128.2])

dt1 = np.dtype([('name', np.unicode_, 16), ('sales', np.float64, (len(games1),))])
dt2 = np.dtype([('name', np.unicode_, 16), ('sales', np.float64, (len(games2),))])

sales_list1 = np.array([('na_Sales', (games1['na_Sales'])), ('eu_Sales', (games1['eu_Sales'])), ('jp_Sales', games1['jp_Sales'])],dtype=dt1)
sales_list2 = np.array([('na_Sales', (games2['na_Sales'])), ('eu_Sales', (games2['eu_Sales'])), ('jp_Sales', games2['jp_Sales'])],dtype=dt2)

g_cost1 = np.array([np.multiply(sales_list1[0]['sales'],g_price1[0]), np.multiply(sales_list1[1]['sales'],g_price1[1]), np.multiply(sales_list1[2]['sales'],g_price1[2])])
g_cost2 = np.array([np.multiply(sales_list2[0]['sales'],g_price2[0]), np.multiply(sales_list2[1]['sales'],g_price2[1]), np.multiply(sales_list2[2]['sales'],g_price2[2])])

print("\n\n-----------------Task 6 & 7------------------------\n\n",g_cost1,'\n\n',g_cost2)





#____________________________Task8______________________________________________

def checkCost():
  print("\n\n-----------------Task8------------------------\n\n")
  other_list1 = np.array([('other_Sales', (games1['other_Sales']))],dtype=dt1)
  other_list2 = np.array([('other_Sales', (games2['other_Sales']))],dtype=dt2)

  temp_sales = sales_list1
  temp_others = other_list1[0][1]

  r1 = 0
  r2 = 0
  
  for i in range(2):
    total_region_sales = sum(temp_sales['sales'])

    k = 0
    for total_sale in total_region_sales:
      if total_sale < temp_others[k]:
        k += 1
        if i == 0:
          r1 += 1
        else: r2 += 1

    temp_sales = sales_list2
    temp_others = other_list2[0][1]

  if r1 > r2:
    print("The total region sales of the first part(r1), are ",str(r1-r2)," more than the second part.")
  elif r2 > r1:
    print("The total region sales of the second part(r1), are ",str(r2-r1)," more than the first part.")
  elif r1 & r2 == 0:
    print("The total region sales of both parts are not less than the other sales.(r1,r2 = 0)")
  else: 
    print("Total sales of ",str(r1)," units of both parts are less than other sales.")

  print("r1: ",r1)
  print("r2: ",r2)

checkCost()
```
