## 8. (Exercise) Сформувати зрізи для отримання зафарбованих елементів.

```python
import numpy as np
matrixzeros = np.zeros((4,4), np.dtype(int))
matrixones_vertical = np.ones((6,1),np.dtype(int))
matrixones_horizontal = np.ones((1,4),np.dtype(int))

matrix = np.concatenate([matrixones_horizontal,matrixzeros], axis=0)
matrix = np.concatenate([matrix,matrixones_horizontal], axis=0)

matrix = np.concatenate([matrixones_vertical,matrix], axis=1)
matrix = np.concatenate([matrix,matrixones_vertical], axis=1)

print(matrix)
print("\n")

first_row = matrix[0]
last_row = matrix[-1]

first_column = matrix[:,0]
last_column = matrix[:,-1]




print("first row: "+str(first_row),"\nLast row: "+str(last_row))
print("\n")
print("First column: " + str(first_column),"\nLast_column: " +str(last_column) + "\n")


deleted_matrix = np.delete(matrix,0,0)
deleted_matrix = np.delete(deleted_matrix,-1,0)
deleted_matrix = np.delete(deleted_matrix,0,1)
deleted_matrix = np.delete(deleted_matrix,-1,1)

print(deleted_matrix)

```

## 8. Сформувати зрізи для отримання зафарбованих елементів.

```python
import numpy as np

numpy_array = np.array([2,3,6,8,11,34])
numpy_matrix = np.array([2,3,6,8,11,34,11,22,3,67,11,87,3,6,3,11,89,14,4,5,8,1,4,67])
numpy_matrix = numpy_matrix.reshape(4,6)
selected_items = []

print(str(numpy_array)+"\n")
print(str(numpy_matrix)+"\n")
print("_______________________\n 1:")

#1
for i in range(2,4):
  selected_items.append(numpy_array[i])
print("\n"+str(selected_items)+"\n")
print("_______________________\n 2:")

#2
print(str(numpy_matrix[:,1].reshape(4,1))+"\n")
print(str(numpy_matrix[:,2].reshape(4,1))+"\n")
print("_______________________\n 3:")

#3
print(str(numpy_matrix[:,1].reshape(4,1))+"\n")
print(str(numpy_matrix[:,3].reshape(4,1))+"\n")
print(str(numpy_matrix[:,5].reshape(4,1))+"\n")
print("_______________________\n 4:")

#4
print(str(numpy_matrix[1:5,4].reshape(3,1))+"\n")
print(str(numpy_matrix[1:5,5].reshape(3,1))+"\n")
print("_______________________\n 5:")

#5
print(str(numpy_matrix[0:2]))
print("_______________________\n 6:")

#6
print(str(numpy_matrix[2,0:3]))
```
