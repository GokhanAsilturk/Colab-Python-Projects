1. Виберіть підмножини даних (кінцевий набір даних (дані штатів США) з
попередньої практичної роботи) на основі стовпця «штат\регіон».
Виведіть кількість спостережень у кожній групі.

```python
from numpy.ma.core import default_fill_value
from google.colab import drive
import pandas as pd
import numpy as np

url_merged_pop_area = '/content/drive/MyDrive/Colab Notebooks/csv_files/merged_pop_area.csv'

merged_pop_area = pd.read_csv(url_merged_pop_area, sep=',')

merged_pop_area = merged_pop_area.drop(merged_pop_area.loc[merged_pop_area['state/region'] == 'PR'].index,axis=0)

print(merged_pop_area.head(),'\n')

all_states = list(merged_pop_area['state'].unique())
all_states_count = {}
for state in all_states:
  all_states_count[state] = list(merged_pop_area['state']).count(state)


print(all_states_count)
```

2. Обчисліть середнє значення населення для кожного штату.

```python
population_total = merged_pop_area.loc[merged_pop_area['ages'] == 'total']
print(population_total.groupby('state')['population'].median())
```

3. Виведіть мінімальне значення населення для кожного штату та року.

```python
print(merged_pop_area.groupby(['state', 'year'])['population'].min())
```

4. Виберіть підмножини даних на основі стовпця «рік». Обчисліть суму

```python
merged_pop_area.loc[(merged_pop_area['state/region'] != 'USA') & (merged_pop_area['ages'] == 'total')].groupby(['year'])['population'].sum()
```

5. За допомогою методу aggregate() виведіть медіану, стандартне та
максимальне значення для кожного штату.

```python
print(population_total.groupby('state')['population'].aggregate([np.median, 'std', max]))
```

6. За допомогою методу filter() виведіть 10 перших груп, у яких
стандартне відхилення сукупності більше за X.

```python
def filt_std(x):
    return x['population'].std() > 1270504

print(population_total.groupby('state/region').filter(filt_std).head(10))
```

7. Використовуючи метод filter() і лямбда-функції, виведіть назву штатів
які брали участь у наборі даних 5 або більше разів.

```python
population_total.groupby('state/region').filter(lambda x: x['state'].value_counts() > 5).groupby('state').groups.keys()
```
