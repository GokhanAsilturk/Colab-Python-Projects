Задати рядок, що містить імена клієнтів, наприклад:
C_names = “Barnes Abramson Macduff Johnson Andrews Lewin Brown”

```python
import numpy as np

c_names = "Barnes Abramson Macduff Johnson Andrews Lewin Brown"

customers_total_statistics = {}
customers_logs = {}

customers_total_amount = {}
customers_monthly_amount = {}


def show_total_statistics(month):
    print("\tStatistics for %d months:\n\n" % month)
    splitted_names = c_names.split()

    for name in splitted_names:
        # find_customers_statistics(name,month)
        find_customers_statistics(name, month)
    for customer, total in customers_total_statistics.items():
        print("customer: %s\nStatistic for %d months: %d minutes\n" % (customer, month, total))


def find_customers_statistics(customer_name, month):
    phoneCall_statistic = np.random.randint(100, 350, size=month)
    customers_logs.setdefault(customer_name, phoneCall_statistic)
    total_statistic = sum(phoneCall_statistic)  # istatistiklerin toplamını alıyoruz
    customers_total_statistics.setdefault(customer_name, total_statistic)


def calculate_amount(dollar_price, course):
    current_price = dollar_price * course

    for name, logArray in customers_logs.items():
        amountList = []
        for log in logArray:
            amountList.append(int(current_price) * int(log))

        amountList_numpy = np.array(amountList)
        customers_monthly_amount.setdefault(name, amountList_numpy)
        total_amount = sum(amountList_numpy)
        customers_total_amount.setdefault(name, total_amount)


# -------------------2-------------------

def info_selected_customer(customer_name):
    customer_name = customer_name.title()

    print("\n\n\n-------------------------------\n\n\ncustomer: %s" % customer_name)

    if customer_name in customers_total_statistics:
        print("\ncustomer logs monthly:")
        i = 1
        for log in customers_logs[customer_name]:
            if i < (len(customers_logs[customer_name]) / 2):
                print(" " + str(i) + " month: %s minutes" % str(log), end=" ")
                i = i + 1
            elif i == (len(customers_logs[customer_name]) / 2):
                print(" " + str(i) + " month: %s minutes" % str(log))
                i = i + 1
            else:
                print(" " + str(i) + " month: %s minutes" % str(log), end=" ")
                i = i + 1
        

        print("\nTotal statistic: %d minutes" % (customers_total_statistics[customer_name]))

    if customer_name in customers_monthly_amount:
        print("\ncustomer monthly amount:")
        i = 1
        for amount in customers_monthly_amount[customer_name]:
            if i < (len(customers_monthly_amount[customer_name]) / 2):
                print(" " + str(i) + " month: %s UAH" % str(amount), end=" ")
                i = i + 1
            elif i == (len(customers_monthly_amount[customer_name]) / 2):
                print(" " + str(i) + " month: %s UAH" % str(amount))
                i = i + 1
            else:
                print(" " + str(i) + " month: %s UAH" % str(amount), end=" ")
                i = i + 1

        print("\nTotal amount: %d UAH / %d$" % (
            customers_total_amount[customer_name], customers_total_amount[customer_name] / float(1.6)))


# -------------------3-------------------


def get_input_for_filter():
    name_list = []
    for number in range(2):
        name_of_customer = input("Enter customer %d: " % (number + 1))
        name_list.append(name_of_customer.title())
    print("\n\n")
    info_without_unselected_customers(name_list)


def info_without_unselected_customers(name_list):
    check = (numpy_customer_names != str(name_list[0])) & (
            numpy_customer_names != str(name_list[1]))  # true false bilgisi alıyoruz

    new_customer_names = numpy_customer_names[check]
    new_customer_logs = numpy_customer_logs[check]
    new_customer_total_logs = numpy_customer_total_logs[check]
    new_customer_monthly_amount = numpy_customer_monthly_amount[check]
    new_customer_total_amount = numpy_customer_total_amount[check]

    for index in range(len(new_customer_names)):
        print("Name: ", new_customer_names[index])

        print("Monthly logs: ", end="")
        for i in range(len(new_customer_logs)):
            if i == len(new_customer_logs) - 1:
                print(str(new_customer_logs[index][i]), "min")
            else:
                print(str(new_customer_logs[index][i]), "min", end=" ")

        print("Total log: ", new_customer_total_logs[index], " minutes")

        print("Monthly amount: ", end="")
        for i in range(len(new_customer_logs)):
            if i == len(new_customer_monthly_amount) - 1:
                print(str(new_customer_monthly_amount[index][i]), "UAH")
            else:
                print(str(new_customer_monthly_amount[index][i]), "UAH", end=" ")

        print("Total amount: ", new_customer_total_amount[index], "UAH\n")


# -------------------4-------------------

def info_for_2_4_months(customer_name):
    customer_name = customer_name.title()

    customer_index = list(numpy_customer_names).index(customer_name)
    customer_log = numpy_customer_logs[customer_index]
    customer_monthly_amount = numpy_customer_monthly_amount[customer_index]

    print("Name: %s" % customer_name)
    print("customer monthly logs: ")
    new_log_list = []
    total_amount = 0
    for i in range(2, 5):
        new_log_list.append(customer_log[i + 1])
        print("%d month: %d min = %d UAH / %d$" % (
            i, customer_log[i + 1], customer_monthly_amount[i + 1], customer_monthly_amount[i + 1] / 1.6))
        total_amount = total_amount + customer_monthly_amount[i + 1]

    print("Total log for 2-4 months: ", sum(new_log_list), "minutes")
    print("Total amount for 2-4 months: %d UAH / %d$" % (total_amount, total_amount / 1.6))


# -------------------5-------------------


def info_for_1_month_without_1_person(customer_name):
    customer_name = customer_name.title()
    check = (numpy_customer_names != customer_name)

    new_customer_names = numpy_customer_names[check]
    new_customer_logs = numpy_customer_logs[check]
    new_customer_total_logs = numpy_customer_total_logs[check]
    new_customer_monthly_amount = numpy_customer_monthly_amount[check]
    new_customer_total_amount = numpy_customer_total_amount[check]

    for index in range(len(new_customer_names)):
        print("Name: ", new_customer_names[index])

        print("Monthly logs: ", end="")
        for i in range(len(new_customer_logs)):
            if i == len(new_customer_logs) - 1:
                print(str(new_customer_logs[index][i]), "min")
            else:
                print(str(new_customer_logs[index][i]), "min", end=" ")

        print("Total log: ", new_customer_total_logs[index], " minutes")

        print("Monthly amount: ", end="")
        for i in range(len(new_customer_logs)):
            if i == len(new_customer_monthly_amount) - 1:
                print(str(new_customer_monthly_amount[index][i]), "UAH")
            else:
                print(str(new_customer_monthly_amount[index][i]), "UAH", end=" ")

        print("Total amount: %dUAH / %d$" % (
            new_customer_total_amount[index], int(new_customer_total_amount[index] / 1.6)) + "\n")


# -------------------6-------------------
def calculate_dollar(price):
    monthly_amounts = []

    for logs in numpy_customer_logs:
        temp_monthly_amount = []
        for log in logs:
            temp_monthly_amount.append(int(log * price))
        monthly_amounts.append(temp_monthly_amount)

    numpy_monthly_amount = np.array(list(monthly_amounts))
    numpy_random_month = np.random.randint(0, len(numpy_monthly_amount) - 1, size=len(numpy_customer_names))
    for name in numpy_customer_names:
        amounts = numpy_monthly_amount[np.where(numpy_customer_names == name)]
        random_month_index = numpy_random_month[np.where(numpy_customer_names == name)]
        print("\ncustomer Name: ", name)
        print("Amount for %s month: " % int(random_month_index + 1), int(amounts[0][random_month_index]), "$")

    dollar_amounts.append(numpy_monthly_amount)


# ------------------7-------------------

def calculate_grn(customer_name, price):
    customer_dollar_amount_list = dollar_amounts[0][np.where(numpy_customer_names == customer_name)]
    total_amount = sum(customer_dollar_amount_list[0]) * price
    print("customer Name: %s" % customer_name.title())
    for dollar_amount in customer_dollar_amount_list[0]:
        print(
            "%d month: %d UAH" % (list(customer_dollar_amount_list[0]).index(dollar_amount) + 1, dollar_amount * price))
    print("Total amount: %d UAH" % total_amount)


# --------------------------------main------------------------------------------
print("\n\n---------------1----------------\n")
show_total_statistics(6)
calculate_amount(float(1.6), int(28))
print("\n")

numpy_customer_names = np.array(list(customers_total_statistics.keys()))
numpy_customer_logs = np.array(list(customers_logs.values()))
numpy_customer_total_logs = np.array(list(customers_total_statistics.values()))
numpy_customer_monthly_amount = np.array(list(customers_monthly_amount.values()))
numpy_customer_total_amount = np.array(list(customers_total_amount.values()))
# -------------------2-------------------
print("\n\n---------------2----------------\n")
info_selected_customer("Abramson")
info_selected_customer("johnson")

# ------------------3-------------------
print("\n\n---------------3----------------\n")
get_input_for_filter()


# ------------------4-------------------
print("\n\n---------------4----------------\n")
info_for_2_4_months('brown')

# ------------------5-------------------
print("\n\n---------------5----------------\n")
info_for_1_month_without_1_person('barnes')

# ------------------6-------------------
print("\n\n---------------6----------------\n")
dollar_amounts = []
calculate_dollar(1.6)

# ------------------7-------------------
print("\n\n---------------7----------------\n")
calculate_grn('Andrews', 28)
```

## Завдання 2. Робота з модулем random, fancy indexing

Тривалість розмов клієнтів мобільного оператора в роумінгу за останні M місяців варіюється від 1000 до 4000 хвилин.
Припустимо, у компанії обслуговується лише N абонентів.
Згенерувати NumPy, що містить інформацію про “Roaming call duration”, використовуючи відповідний метод модуля random.
Використовуючи Fancy indexing вибрати підмножину у вказаному порядку, представленому двома 2 NumPy масивами:

i=[[2,4],[7,8]], 

j=[[0,2],[9,6]]

Варіант: 3

N=31 M=62

```python
import numpy as np


def create_random_logs(month,number_of_customers):
  
  customers_call_duration = np.random.randint(1000,4001, [number_of_customers, month])
  
  i = np.array([2, 4, 7, 8]).reshape(2,2)
  j = np.array([0, 2, 9, 6]).reshape(2,2)

  print(customers_call_duration, customers_call_duration[i , j] , sep='\n\n')



create_random_logs(62,31)



```
