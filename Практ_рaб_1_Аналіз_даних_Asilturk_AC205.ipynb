Комбінація з двох символів "\t " призводить до появи табуляції в рядку. Вкладки можна використовувати в будь-якому місці рядка, де вам

## Доступ до одного елемента в списку

* Елементи в списку визначаються їх позицією в списку, починаючи з нуля.
* Число в дужках називається індексом елемента.

наприклад:

```python
city_list = ['antalya','istanbul','ankara','alanya']

city = city_list[0]
print('Selected city: ' + city.title())

city = city_list[2]
print('\nSelected city: ' + city.title())
```

```python
city_list = ['antalya','istanbul','ankara','alanya']

city = city_list[-1]

print(city.title())
```

# **Списки та цикл**

##Доступ до всіх елементів у списку

Ми використовуємо цикл для доступу до всіх елементів списку. Цикл – це блок коду, який повторюється, доки в ньому не закінчаться елементи для роботи або доки не буде виконана певна умова. У цьому випадку наш цикл буде виконуватися один раз для кожного елемента нашого списку. Зі списком, який складається з трьох елементів, наш цикл буде виконуватися тричі.

```python
product_brands = ['Apple','Samsung','LG','Philips','Asus']

for brand in product_brands:
  print('Brands filtered: ' + brand)
```

* Ключове слово «for» повідомляє Python підготуватися до використання циклу.

* Змінна "product" є тимчасовою змінною-заповнювачем. Це
змінної, в яку Python поміщає кожен елемент зі списку по одному.

* Під час першого проходження циклу значенням "product" буде "Apple".
* Під час другого циклу значення «dog» буде «Samsung».
* Утретє "product" буде «LG».
* У четвертий раз "product" буде «Philips».
* У п’ятий раз "product" буде «Asus».
* Після цього в списку більше немає елементів, і цикл закінчиться.


## Перерахування списку

Функція enumerate() відстежує індекс кожного елемента для вас, перебираючи список:

```python
brands = ['Apple','Samsung','LG','Philips','Asus']

print("Перелік тарифів продажу виглядає наступним чином:\n")

for index, brand in enumerate(brands):
  #Значення в індексі змінної завжди є цілим числом.
  #Якщо ви хочете надрукувати його в рядку, 
  #вам потрібно перетворити ціле число в рядок:
  current_place = str(index + 1)
  print("Place: " + current_place + " Brand: " + brand)
```
