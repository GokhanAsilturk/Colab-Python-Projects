1. Використовуючи функцію pd.read_csv прочитати з файлу відповідну частину стовбців name, generation, type1, type2, hp, speed, sp_attack, sp_defense (діапазон строк задається по варіанту студента, див. ДодатокA) у DataFrame pokemon. Стовбець abilities перетворити у 2 стовбця abilities1, abilities2, розподілити значення відповідно.

Діапазон (від): 717

Діапазон (до, включаючи): 485

N1:5 N2:52 M1:10 M2:46

```python
from google.colab import drive
import pandas as pd
import numpy as np

url = '/content/drive/MyDrive/Colab Notebooks/csv_files/pokemon.csv'

pokemon_selected = pd.read_csv(url, usecols = [30, 39, 36, 37, 28, 35, 33, 34], skiprows = 484, nrows = 717-484, header=None)
pokemon_selected = pd.DataFrame(pokemon_selected[::-1])

pokemon_abilities = pd.read_csv(url, usecols = [0], skiprows = 484, nrows = 717-484, header=None)
pokemon_abilities = pd.DataFrame(pokemon_abilities[::-1])
pokemon_abilities = pokemon_abilities[0].str.split(',', expand = True)

pokemon_final = pd.concat([pokemon_abilities, pokemon_selected],axis=1).drop([2, 3], axis=1)
pokemon_final.columns = ['ability1', 'ability2', 'hp', 'name', 'sp_attack', 'sp_defense', 'speed', 'type1', 'type2', 'generation']
pokemon_final = pokemon_final.reindex(columns = ['name', 'generation', 'type1', 'type2', 'hp', 'speed', 'sp_attack', 'sp_defense', 'ability1', 'ability2'])

print(pokemon_final)
```
