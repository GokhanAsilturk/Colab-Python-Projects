```python
sellers_models = {
    "Apple Offical": ['7','8','X'],
    "rozetka.com" : ['6','7','7 Plus','8','8 Plus','X','11'],
    "citrus.com" : ['6s','7','11','12'],
    "amazon.com" : ['6','7 Plus','8','13 Pro Max']
}


models_prices = {
    '5' : 200,
    '6' : 300,
    '6s' : 350,
    '7' : 400,
    '7 Plus' : 480,
    '8' : 550,
    '8 Plus' : 620,
    'X' : 700,
    '11' : 800,
    '12' : 930,
    '13 Pro Max' : 1100,
    
}

prices = []
sorting_prices = []
sorting_models = []
def show_phones(out_of_model,target_price):
  print("\tResults of Search\n\n")
  print("filtered -> lowest model: %s+ , max price: %d$" % (out_of_model,target_price))

  for seller, models in sellers_models.items(): 
    stock_info = ""
    
    
    for model in models:
      
      
      for model_type, price in models_prices.items():
        if model_type==model and check_price(price,target_price) and check_model(model,out_of_model):
          stock_info = stock_info + "\n" + ('Iphone %s : %d$' % (model,price))
          sorting_prices.append(price)
          sorting_models.append(model)
      
    if stock_info != "":
      print("Seller: %s" %seller)
      print(stock_info + "\n\n")
          
        

    print('\n')
  
      

def check_price(price,target_price):

  if price <= target_price:
    return True
  else:
    return False


def check_model(model,out_of_model):

  if model != out_of_model:
    return True
  else:
    return False



def sort_by_price(sorting_prices,sorting_models):
  print("Prices are sorted -> from high to low\n")
  
  sorting_prices.sort(reverse = True)
  
  sorted_items = {}
  
  for sorted_price in sorting_prices:
    
    for model,price in models_prices.items():
      if models_prices[model] == sorted_price:
        sorted_items[model] = price


  for sorted_model,sorted_price in sorted_items.items():

    for seller,seller_models in sellers_models.items():

      for model in seller_models:
        if model == sorted_model:
          print('Iphone %s : %d$' % (sorted_model,sorted_price))
          print("Seller: %s\n" %seller)

show_phones('6',500)

sort_by_price(sorting_prices,sorting_models)
```
